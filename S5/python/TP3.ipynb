## EXERCICE 1


```python
# puissance

def puissance(a, b):
    if b == 0:
        return 1
    return a * puissance(a, b - 1)

print(puissance(5, 3))

# calcul factoriel

def fact(n):
    if n == 1:
        return 1
    return n * fact(n - 1)

print(fact(5))
# calcul des fibbonacci

def fibbonacci(nombre):
    if nombre <= 2:
        return 1
    return fibbonacci(nombre - 2) + fibbonacci(nombre - 1)

print(fibbonacci(100))

# PGCD
def pgcd(a, b):
    if b == 0:
        return a
    else:
        Rest = a % b
        return pgcd(b, Rest)

print(pgcd(20, 40))

# somme d'elements de liste

def somme_liste(lst):
    if not lst:
        return 0
    return lst[0] + somme_liste(lst[1:])
        
liste = [1, 2, 3, 5]
print(somme_liste(liste))

# calculer la somme des chiffres nombre entier

def somme_chiffre(n):
    if n == 0:
        return 0
    digit = n % 10
    return digit + somme_chiffre(n // 10)

print(somme_chiffre(2235))

# comptage d'occurence 

def nombre_occurrence(lst, n):
    if not lst:
        return 0
    if lst[0] == n:
        return 1 + nombre_occurrence(lst[1:], n)
    return 0 + nombre_occurrence(lst[1:], n)

liste = [1, 2, 3, 5, 3]
print(nombre_occurrence(liste, 3))

```

# EXERCICE 3 :


```python
ventes = {"Dupont": 14, "Hervy": 19, "Geoffroy": 35, "Layee": 21}

def nombre_total(dct):
    return sum(dct.values())

print(nombre_total(ventes))

def plus_vendeur(dct):
    maximum = max(dct.values())
    for key, value in dct.items():
        if value == maximum:
            return {key: value}

print(plus_vendeur(ventes))

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100)
y = np.cos(x) + 3 * np.sin(2 * x)

plt.plot(x, y, label='Y = cos(x) + 3sin(2x)', color='red')
plt.xlabel('x')
plt.ylabel('Y')

plt.grid(True)
plt.legend()

plt.show()
```
