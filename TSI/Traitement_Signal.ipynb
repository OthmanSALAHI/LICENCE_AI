##  UN SINUSOIDAL

```python
import numpy as np
import matplotlib.pyplot as plt

# les parametres de signal
A = 1 # Amplitude
f = 5 # frequence du signal (Hz)
fs = 20 # frequence de d'chontillonnage
duree = 5
n = np.arange(0, duree, 1/fs) # Instance discrets (de 0  1 s, avec un pas de 1/fs)

# signal du sinosoidal discret
signal_discret = A * np.sin(2 * np.pi * f * n)

# AFFICHAGE
plt.stem(n, signal_discret, basefmt=" ") # supprimer 'use line collection'
plt.title("Singal sinosoidal discret")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# les parametres de signal
fs = 500 # frequence de d'chontillonnage
f = 5 # frequence du signal (Hz)
duration = 1 # dure en  secondes
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# signal du sinosoidal discret
signal = 2 * np.sin(2 * np.pi * f * t)

# AFFICHAGE
plt.stem(t, signal) # supprimer 'use line collection'
plt.title("Singal sinosoidal discret")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft  # Import fft from scipy.fft

# Paramtres du signal
A = 1  # Amplitude
f = 5  # Frquence du signal (Hz)
fs = 100  # Frquence d'chantillonnage (doit tre au moins > 2*f selon Shannon)
t = np.arange(0, 1, 1/fs)  # Instants discrets (0  1s avec un pas de 1/fs)

# Gnrer le signal sinusodal
signal = A * np.sin(2 * np.pi * f * t)

# 1 Affichage du signal temporel
plt.figure(figsize=(10, 4))
plt.stem(t, signal)  # Ajout de use_line_collection pour viter le warning
plt.title("Signal sinusodal discret")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

# 2 Calcul et affichage de la FFT
N = len(signal)  # Nombre d'chantillons
fft_signal = fft(signal)  # Calcul de la FFT
frequencies = np.fft.fftfreq(N, 1/fs)  # Axe des frquences

# On ne garde que la moiti des valeurs (FFT est symtrique pour un signal rel)
half_N = N // 2
frequencies = frequencies[:half_N]
fft_magnitude = np.abs(fft_signal[:half_N])  # Magnitude du spectre

plt.figure(figsize=(10, 4))
plt.stem(frequencies, fft_magnitude)
plt.title("Spectre de Fourier du signal")
plt.xlabel("Frquence (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
