##  UN SINUSOIDAL

```python
import numpy as np
import matplotlib.pyplot as plt

# les parametres de signal
A = 1 # Amplitude
f = 5 # frequence du signal (Hz)
fs = 20 # frequence de d'échontillonnage
duree = 5
n = np.arange(0, duree, 1/fs) # Instance discrets (de 0 à 1 s, avec un pas de 1/fs)

# signal du sinosoidal discret
signal_discret = A * np.sin(2 * np.pi * f * n)

# AFFICHAGE
plt.stem(n, signal_discret, basefmt=" ") # supprimer 'use line collection'
plt.title("Singal sinosoidal discret")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft  # Import fft from scipy.fft

# Paramètres du signal
A = 1  # Amplitude
f = 5  # Fréquence du signal (Hz)
fs = 100  # Fréquence d'échantillonnage (doit être au moins > 2*f selon Shannon)
t = np.arange(0, 1, 1/fs)  # Instants discrets (0 à 1s avec un pas de 1/fs)

# Générer le signal sinusoïdal
signal = A * np.sin(2 * np.pi * f * t)

# 1️⃣ Affichage du signal temporel
plt.figure(figsize=(10, 4))
plt.stem(t, signal)  # Ajout de use_line_collection pour éviter le warning
plt.title("Signal sinusoïdal discret")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

# 2️⃣ Calcul et affichage de la FFT
N = len(signal)  # Nombre d'échantillons
fft_signal = fft(signal)  # Calcul de la FFT
frequencies = np.fft.fftfreq(N, 1/fs)  # Axe des fréquences

# On ne garde que la moitié des valeurs (FFT est symétrique pour un signal réel)
half_N = N // 2
frequencies = frequencies[:half_N]
fft_magnitude = np.abs(fft_signal[:half_N])  # Magnitude du spectre

plt.figure(figsize=(10, 4))
plt.stem(frequencies, fft_magnitude)
plt.title("Spectre de Fourier du signal")
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
