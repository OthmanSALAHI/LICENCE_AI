# **Load data**

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv")
df
```

# **data preparation**

### data separation x and y ###

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

### splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

the train data take 80% of dataset , otherwise the test data take 20% of dataset

```python
x_train
```

# **model building**

### linear regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

#### **training model**

```python
lr.fit(x_train, y_train)
```

#### applying the model to make the prediction

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

 #### Evaluate model performance


```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('the mean squared error of linear regression (train):')
lr_train_mse
```

```python
print('the R2 score of Linear regression (train):')
lr_train_r2
```

```python
print('the mean squared error of Linear regression (test):')
lr_test_mse
```

```python
print('the R2 score of Linear regression (test):')
lr_test_r2
```

```python
lr_result = pd.DataFrame(["Linear regression", lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_result.columns = ["methode","Training_MSE", "Training_R2_score", "testing_MSE", "Testing_R2_score"]
lr_result
```

### **Random forest**

#### training the model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

#### applying the model to make the prediciton

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
y_rf_train_pred
```

 #### Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_result = pd.DataFrame(["Random Forest", rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_result.columns = ["methode","Training_MSE", "Training_R2_score", "testing_MSE", "Testing_R2_score"]
rf_result
```

### **model comparasion**

```python
df_models = pd.concat([lr_result, rf_result], axis=0).reset_index(drop=True)
df_models
```
