```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import f_oneway, chi2_contingency

np.random.seed(42)

data = {
    'ID': range(1, 201),
    'Filière': np.random.choice(['Sciences', 'Lettres', 'Économie', 'Informatique'], 200),
    'Note_Math': np.clip(np.random.normal(12, 3, 200), 0, 20),
    'Note_Litt': np.clip(np.random.normal(14, 2.5, 200), 0, 20),
    'Âge': np.random.randint(18, 26, 200),
    'Présence': np.random.uniform(40, 100, 200),
    'Sexe': np.random.choice(['M', 'F'], 200)
}

etudiants = pd.DataFrame(data)

print("Aperçu des 5 premières lignes :")
print(etudiants.head())

print("\nRésumé statistique :")
print(etudiants.describe())
```

```python
sns.set_style("whitegrid")

sns.pairplot(etudiants, hue='Filière', diag_kind='kde',
             vars=['Note_Math', 'Note_Litt', 'Âge', 'Présence'],
             palette='Set2', markers=['o', 's', 'D', '^'])

plt.suptitle('Relations entre variables par filière', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Filière', y='Note_Math', hue='Sexe', data=etudiants, palette='Pastel1')

plt.title('Répartition des notes en mathématiques par filière et sexe')
plt.xlabel('Filière')
plt.ylabel('Note en mathématiques (0-20)')
plt.legend(title='Sexe')

plt.show()
```
